4. Check if $R$ is a reflexive relation.

   A relation R on a set A is reflexive if and only if for every element a in A, there exists an ordered pair (a,a) in R.

```python
def is_reflexive(A, R):
    for a in A:
        if (a, a) not in R:
            return False
    return True

A = {1, 2, 3, 4}
B = {'x', 'y', 'z'}
R = {(1, 'y'), (1, 'z'), (3, 'y'), (4, 'x'), (4, 'z')}

result = is_reflexive(A, R)
print(f"R is a reflexive relation? {result}")  

import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_nodes_from(A)
G.add_nodes_from(B)
G.add_edges_from(R)

nx.draw(G, with_labels=True)
plt.show()


```

5. Check if $R$  is a symmetric relation.

   A relation R is symmetric if and only if for every pair (a,b) in R, the pair (b,a) is also in R.

```python
def is_symmetric(R):
    for a, b in R:
        if (b, a) not in R:
            return False
    return True

R = {(1, 'y'), (1, 'z'), (3, 'y'), (4, 'x'), (4, 'z')}

print(f"R is a symmetric relation? {is_symmetric(R)}")

import networkx as nx
import matplotlib.pyplot as plt

R = {(1, 'y'), (1, 'z'), (3, 'y'), (4, 'x'), (4, 'z')}

G = nx.DiGraph()
G.add_edges_from(R)

nx.draw(G, with_labels=True)
plt.show()

```
